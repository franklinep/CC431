```python
# Replicamos lo realizado en clase
img_roja  = imread('Maquina-Filtro-Rojo.jpeg')
img_verde = imread('Maquina-Filtro-Verde.jpeg')
img_azul = imread('Maquina-Filtro-Azul.jpeg')

# Las medidas de nuestra imagen
alto, ancho, capas = img_roja.shape
print("El tamaño de la imagen es: ")
print("Alto: ", alto, "px\nAncho:", ancho, "px\nCapas:", capas, "px")

# Mostramos las imagenes de distinto filtro
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1); plt.imshow(img_roja); plt.title('Filtro Rojo')

plt.subplot(1, 3, 2); plt.imshow(img_verde); plt.title('Filtro Verde')

plt.subplot(1, 3, 3); plt.imshow(img_azul); plt.title('Filtro Azul')

plt.show()
```

```python
# Mostramos la IMAGEN DE FILTRO ROJO en distintas canales de color.

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1); plt.imshow(img_roja[:,:,0], cmap='gray'); plt.title('Canal "R"')

plt.subplot(1, 3, 2); plt.imshow(img_roja[:,:,1], cmap='gray'); plt.title('Canal "G"')

plt.subplot(1, 3, 3); plt.imshow(img_roja[:,:,2], cmap='gray'); plt.title('Canal "B"')

plt.show()
```

# ¿Por qué no se ve igual en las mismas capas?

Observamos una imagen de filtro rojo, donde sus valores para las capas de color Verde y Azul
son bajos a diferencia del Rojo, por lo que en escala de grises estos se ven distintos.

```python
# Mostramos la imagen completa con las imagenes de filtro, recordando el orden RGB

parte_roja  = img_roja[:,:,0]  # R
parte_verde = img_verde[:,:,1] # G
parte_azul  = img_azul[:,:,2]  # B

imagen_completa = np.stack([parte_roja, parte_verde, parte_azul], axis=-1)
plt.imshow(imagen_completa)
```

# ¿Qué pasaría si se tomara otras partes de las imágenes filtradas?
Si se toman diferentes partes de las imágenes filtradas para formar la imagen completa el resultado seria distinto al original, ya que estariamos intercambiando los canales de color entre las diferentes imagenes.


```python
parte_roja  = img_roja[:,:,2]  # R
parte_verde = img_verde[:,:,0] # G
parte_azul  = img_azul[:,:,1]  # B

imagen_completa = np.stack([parte_roja, parte_verde, parte_azul], axis=-1)
plt.imshow(imagen_completa)
```
